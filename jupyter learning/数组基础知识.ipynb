# 2 array的基本操作

## 2.1 生成01数组

```python
ones = np.ones([2,3,3])
print(ones)
```

```python
zeros = np.zeros([3,3,3])
print(zeros)
```

## 2.2 根据现有数组生成新的数组

```python
x = ['wanghuizhen','jiaxutao','liucheng','wangyansong']
y = np.array(x)
y
```

## 2.3 创建数组的索引

```python
y1 = np.asarray(x)
y1
```

```python
x[0] = ''
print(y)
print(y1)
```

## 2.4 创建固定范围的数组

```python
x = np.linspace(0,100,50,endpoint=True
               )
print(x)
```

## 2.5 创建等差数列

```python
ar_array = np.arange(0,10,2)
print(ar_array)
```

## 2.6 创建等比数列

```python
log_array = np.logspace(0,4,3)
# start = 10^start, stop = 10^end, num = num
print(log_array)
```

## 2.7 生成正态分布

```python
Normal = np.random.randn(100)
# ()character = num
print(Normal)
```

```python
normal = np.random.normal(0, 1, 100000000)
# 均值， 标准差， 数量
print(normal)
```

# 3 画正态分布图 

```python
import matplotlib.pyplot as plt # 导入画图库
```

```python
# 创建画布
plt.figure(figsize=(20,20),dpi=100)
# 绘制直方图
plt.hist(normal,1000)
# first character is data, second is 区间数
plt.show
```
