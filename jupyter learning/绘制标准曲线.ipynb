```python
#å¤é¡¹å¼æå
readData = pd.read_csv("Elisa/TNF-Î±.csv",header = None)
print(readData)
datax = readData.loc[0:7,0].tolist() 
datay = readData.loc[0:7,1].tolist()
y = list(datay) #è¾å¥èªå·±çå¼
x = list(datax) #è¾å¥èªå·±çå¼
print(x,y)
xx = list(range(1,))  #xxå¯ä»¥è®¾ç½®å¤§äºxå¯ä»¥ç¨æ¥åé¢æµ
z1 = np.polyfit(x, y, 6) # ç¨6æ¬¡å¤é¡¹å¼æåï¼å¯æ¹åå¤é¡¹å¼é¶æ°ï¼
p1 = np.poly1d(z1) #å¾å°å¤é¡¹å¼ç³»æ°ï¼æç§é¶æ°ä»é«å°ä½æå
print(p1)  #æ¾ç¤ºå¤é¡¹å¼
yvals=p1(xx) # å¯ç´æ¥ä½¿ç¨yvals=np.polyval(z1,xxx)
plt.plot(x, y, '*',label='original values')
plt.plot(xx, yvals, 'r',label='polyfit values')
plt.xlabel('concentration of protein')
plt.ylabel('OD')
plt.legend(loc=4) # æå®legendå¨å¾ä¸­çä½ç½®ï¼ç±»ä¼¼è±¡éçä½ç½®
plt.title('polyfitting')
plt.show()
```
