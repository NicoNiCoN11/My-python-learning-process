```python
# 创建26*26的DataFrame，行标签为26个大写字母，列标签为26个数字。内容为随机数
import numpy as np
array = np.random.randint(0, 100, (26, 26))
list_mix = []
for i in range(97, 123):
    list_mix.append(chr(i))
list_max = []
for i in range(65, 91):
    list_max.append(chr(i))

my_Frame = pd.DataFrame(array, index = list_max, columns = list_mix)

my_Frame
```
